```python
import os
from random import sample
import cv2
import imutils
import matplotlib.pyplot as plt
%matplotlib inline
import dlib
import numpy as np
```

## Main

```python
# Load face detector
detector = dlib.get_frontal_face_detector()

# Load facial landmarks predictor
# http://dlib.net/files/shape_predictor_68_face_landmarks.dat.bz
predictor = dlib.shape_predictor("../assets/shape_predictor_68_face_landmarks.dat")
```

```python
# Load images.
image = cv2.imread('../data/gabriel.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(image)
```

```python
# Detect face and predict facial landmarks
rects = detector(gray, 1)
rect = rects[0]
shape = predictor(gray, rect) # the landmarks are here
```

```python
# Draw Facial Landmarks

for i in range(shape.num_parts):
    cv2.circle(image, (shape.part(i).x, shape.part(i).y), 4, (0, 255, 0), -1)
    
plt.imshow(image)
```
